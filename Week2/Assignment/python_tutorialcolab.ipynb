```python
from astropy.io import fits
```

```python
img_name ="pdi_pi_collapsed.fits"
img = fits.getdata(img_name, ext=1) # data importing
```

```python
fig = plt.figure(figsize=(8,8)) # create a new figure

image = plt.imshow(img, origin="lower", vmin = -0.001, vmax=0.01)
#The values in the image are colored in these intervals. ( vmin, vmax)
cbar = fig.colorbar(image, shrink=0.82) # color scale bar
cbar.set_label("Surface Brightness (arb.units)", rotation=90, fontsize=12) # color scala label
plt.xlim([60,140]) # values range for x-axis
plt.ylim ([60,140]) # values range for y-axis
plt.xlabel(r'$\Delta$ RA [px]', fontsize=12) # label of x-axis
plt.ylabel(r'$\Delta$ Dec [px]', fontsize=12); # label of y-axis
```
